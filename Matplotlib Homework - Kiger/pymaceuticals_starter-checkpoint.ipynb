## Observations and Insights 

```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
import numpy as np
from scipy.stats import linregress

# Study data files
mouse_metadata_path = "data/Mouse_metadata.csv"
study_results_path = "data/Study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)

# Combine the data into a single dataset
combined_study_df=pd.merge(mouse_metadata,study_results,how='outer',on='Mouse ID')

# Display the data table for preview
combined_study_df
```

```python
# Create a clean DataFrame by dropping the duplicate mouse by its ID.
clean_mouse=combined_study_df.drop_duplicates()
```

## Summary Statistics

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen
# Use groupby and summary statistical methods to calculate the following properties of each drug regimen: 
# mean, median, variance, standard deviation, and SEM of the tumor volume. 
# Assemble the resulting series into a single summary dataframe.

#groupby drug regimen and Tumor volume for statistics using cleaned df
drug_group=clean_mouse.groupby('Drug Regimen')['Tumor Volume (mm3)'].mean().to_frame('Mean')
drug_group['Median']=clean_mouse.groupby('Drug Regimen')['Tumor Volume (mm3)'].median()
drug_group['Variance']=clean_mouse.groupby('Drug Regimen')['Tumor Volume (mm3)'].var()
drug_group['Standard Deviation']=clean_mouse.groupby('Drug Regimen')['Tumor Volume (mm3)'].std()
drug_group['SEM']=clean_mouse.groupby('Drug Regimen')['Tumor Volume (mm3)'].sem()

drug_group.head()
```

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen

# Using the aggregation method, produce the same summary statistics in a single line
drug_group_agg = clean_mouse.groupby('Drug Regimen').agg({'Tumor Volume (mm3)':['mean','median','var','std','sem']})
drug_group_agg.head()
```

## Quartiles, Outliers and Boxplots

```python
# Calculate the final tumor volume of each mouse across four of the treatment regimens:  
# Capomulin, Ramicane, Infubinol, and Ceftamin


# Start by getting the last (greatest) timepoint for each mouse


# Merge this group df with the original dataframe to get the tumor volume at the last timepoint
```

```python
# Calculate the final tumor volume of each mouse across four of the treatment regimens:  
# Capomulin, Ramicane, Infubinol, and Ceftamin
# Start by getting the last (greatest) timepoint for each mouse
max_timepoint = pd.DataFrame(clean_mouse.groupby('Mouse ID')['Timepoint'].max().sort_values()).reset_index().rename(columns={'Timepoint':'Timepoint_max'})
max_timepoint.head()
```

```python
# Merge this group df with the original dataframe to get the tumor volume at the last timepoint
merge_timepoint=pd.merge(clean_mouse,max_timepoint,on='Mouse ID')
merge_timepoint.head(10)
```

```python
# Calculate the final tumor volume of each mouse across four of the treatment regimens:  
# Capomulin, Ramicane, Infubinol, and Ceftamin

# Put treatments into a list for for loop (and later for plot labels)
treatment_list = ['Capomulin', 'Ramicane', 'Infubinol', 'Ceftamin']


# Create empty list to fill with tumor vol data (for plotting)
tumor_vol =[]


#create loop for each drug
for treatment in treatment_list:
    
    # Locate the rows which contain mice on each drug and get the tumor volumes
    #locating rows for treatments in treatment_list
    treatment_df =merge_timepoint.loc[merge_timepoint['Drug Regimen']==treatment]
    
    #use prior greatest timepoint to filter out max volume lines
    tumor_volumes = treatment_df.loc[treatment_df['Timepoint']==treatment_df['Timepoint_max']]
    
    # add subset 
    #find final volume for each drug by appending volumes with treatment list
    volumes = tumor_volumes['Tumor Volume (mm3)']
    tumor_vol.append(volumes)
    
    # Calculate the IQR and quantitatively determine if there are any potential outliers.
    #q_tiles= volumes.quartile([.25,.5,.75])
    quartiles = tumor_volumes['Tumor Volume (mm3)'].quantile([0.25, 0.50, 0.75])
    lower_q=quartiles[0.25]
    upper_q=quartiles[0.75]
    iqr = upper_q - lower_q
    
    # Determine outliers using upper and lower bounds
    lower_bound = lower_q - (1.5*iqr)
    upper_bound = upper_q + (1.5*iqr)   
    
    #outliers=(volumes.loc[(tumor_volumes['Tumor Volume (mm3)'] >= upper_bound) | 
                                        #(tumor_volumes['Tumor Volume (mm3)'] <= lower_bound)]).count()
    
    
    print(f"The lower quartile of {treatment} is: {lower_q}")
    print(f"The upper quartile of {treatment} is: {upper_q}")
    print(f"The interquartile range of {treatment} is: {iqr}")
    print(f"Values below {lower_bound} could be outliers.")
    print(f"Values above {upper_bound} could be outliers.")
   
    
    
    
    
```

```python
# Generate a box plot of the final tumor volume of each mouse across four regimens of interest
fig1, ax1 = plt.subplots()
ax1.set_title('Final Tumor Volume per Drug Regimen')
ax1.set_ylabel('Tumor Volume (mm3)')
ax1.set_xlabel('Drug Regimen')
ax1.boxplot(tumor_vol)
plt.xticks ([1,2,3,4],treatment_list)
plt.show()
```

## Line and Scatter Plots

```python
# Generate a line plot of tumor volume vs. time point for a mouse treated with Capomulin
#locate mouse treated with Capomulin
capomulin_mouse =clean_mouse.set_index('Drug Regimen')
#capomulin_mouse

capomulin= capomulin_mouse.loc['Capomulin',:]
capomulin
```

```python
# Generate a line plot of tumor volume vs. time point for a mouse treated with Capomulin
#prior cell shows Mouse s185 treated with Capomulin
s185=capomulin.loc[capomulin['Mouse ID']=='s185']
plt.plot(s185['Timepoint'], s185['Tumor Volume (mm3)'])
plt.xlabel('Timepoint')
plt.ylabel('Tumor Volume (mm3)')
plt.title(f'Capomulin treated Mouse s185: Timepoint vs. Tumor Vol')
plt.show()
```

```python
# Generate a scatter plot of average tumor volume vs. mouse weight for the Capomulin regimen
#use capomulin dataframe from prior cells for capomulin regimen calculations

#groupby Mouse ID and Tumor volumes to average per mouse ID
capomulin_volumes = pd.DataFrame(capomulin.groupby('Mouse ID')['Tumor Volume (mm3)'].mean()).rename(columns={'Tumor Volume (mm3)':'Avg_Tumor_Volume'})
#capomulin_volumes.head()

#merge capomulin_volumes df with capomulin dataframe to match averages to weights
average_vol = pd.merge(capomulin,capomulin_volumes, on='Mouse ID')
average_vol.head()

```

```python
# Generate a scatter plot of average tumor volume vs. mouse weight for the Capomulin regimen

#remove duplicate lines
weight_vol=average_vol.drop_duplicates()
#weight_vol

plt.scatter(weight_vol['Weight (g)'], weight_vol['Avg_Tumor_Volume'])
plt.xlabel(f'Mouse Weight (g)')
plt.ylabel(f'Average Tumor Volume (mm3)')
plt.title(f'Capomulin Regimen: Tumor Vol vs. Mouse Weight')
plt.show()
```

## Correlation and Regression

```python
# Calculate the correlation coefficient and linear regression model 
# for mouse weight and average tumor volume for the Capomulin regimen

#assign x and y
x_values = weight_vol['Weight (g)']
y_values = weight_vol['Avg_Tumor_Volume']

# Calculate the correlation coefficient
correlation = st.pearsonr(x_values,y_values)
print(f"The correlation between weight and tumor volume for mice treated with Capomulin is {round(correlation[0],2)}")


(slope, intercept, rvalue, pvalue, stderr) = linregress(x_values, y_values)
regress_values = x_values * slope + intercept
line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))
plt.scatter(x_values,y_values)
plt.plot(x_values,regress_values,"r-")
plt.annotate(line_eq,(6,10),fontsize=15,color="red")
plt.xlabel(f'Mouse Weight (g)')
plt.ylabel(f'Average Tumor Volume (mm3)')
plt.title(f'Capomulin Regimen: Tumor Vol vs. Mouse Weight')
plt.show()
```
